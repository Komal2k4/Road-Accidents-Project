**1.Data Preparation & Cleaning**

```python
#Load required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly as px
```

```python
#Missing values
print("Missing values:")
df.isnull().sum()
```

The dataset consists of 3000 rows and 22 columns with both numerical and categorical features. While most fields are complete, some columns such as Traffic Control Presence(716) and Driver License Status(975) contain missing values that need to be handled before analysis.

```python
#Convert time of day to datetime
df['Time of Day']=pd.to_datetime(df['Time of Day'],format="%H:%M",errors="coerce").dt.time
df['Time of Day']
```

```python
#Create Date column
df['Date']=pd.to_datetime(df['Year'].astype(str) + "-" + df['Month'] + "-01",errors="coerce")
df['Date']
```

Create a new 'Date' column by combining 'Year' and 'Month' Convert them into a proper datetime format (YYYY-MM-01) 'errors="coerce"' ensures invalid dates are set to NaT instead of raising errors.

```python
#Make month ordered correctly
months_order=["January","February","March","April","May","June","July","August","September","October","November","December"]
df['Month']=pd.Categorical(df['Month'],categories=months_order,ordered=True)
```

```python
df.sort_values('Month')
```

**Clean missing / invalid values**

```python
placeholders=["Unknown","None","none","NONE"," ","","Expired"]
df=df.replace(placeholders,np.nan)
df.isnull().sum().sort_values(ascending=False)
```

```python
df['City Name']=df['City Name'].fillna("Unknown City")
df['Driver License Status']=df['Driver License Status'].fillna("Not Known")
df['Traffic Control Presence']=df['Traffic Control Presence'].fillna("Unknown")
df.isnull().sum().sort_values(ascending=False)
```

All placeholder text values such as "Unknown", "None", and "Expired" were replaced with NaN to accurately represent missing data. Later, important categorical columns were imputed with appropriate labels like "Unknown City" and "Not Known" to preserve data for visualization.

```python
df=df[df['Driver Age']>15]
```

Only rows with Driver Age > 15 are considered, since drivers below 16 are not legally valid. Others were removed as invalid data.

```python
df=df[df['Speed Limit (km/h)']<201]
```

Only rows with Speed Limit (km/h) < 201 are considered, since in real-world Speed limit > 200 doesn't exists.They were removed as invalid data.

```python
print("Dataset after cleaning:",df.shape)
df.isnull().sum()
```

**Visualization: Accidents per Month**

```python
#Calculate monthly accident counts
month_counts=df['Month'].value_counts().reindex(df['Month'].cat.categories)
month_counts
```

```python
#Get risks
high_risk=month_counts.quantile(0.66)
low_risk=month_counts.quantile(0.33)

#Create colors list based on risk
colors=[]
for i in month_counts:
  if i>high_risk:
    colors.append('red')
  elif i<=low_risk:
    colors.append('green')
  else:
    colors.append('yellow')
```

```python
#plot with our custom colors
plt.figure(figsize=(12,5))
ax=sns.barplot(x=month_counts.index,y=month_counts.values,palette=colors)
plt.title("Accidents per Month in India (Risk Categorized)")
plt.xlabel("Accidents per Month")
plt.ylabel("Number of Accidents")
plt.xticks(rotation=45)

#Add number labels above bars
for bar in ax.patches:
  ax.annotate(format(int(bar.get_height()),'d'),
              (bar.get_x() + bar.get_width()/2,bar.get_height()),
              ha='center',va='bottom')

plt.show()
```

Months are categorized into high, medium, and low accident risk groups based on the number of incidents. High-risk months are marked in red, medium-risk months in yellow, and low-risk months in green. This improves visual understanding of seasonal accident patterns.

**Visualization: Accidents by Speed Limit (Risk Level Based)**

```python
def speed_category(speed):
  if speed<=40:
    return "Low speed"
  elif speed <=80:
    return "Medium speed"
  else:
    return "High speed"
df['Speed Risk Category']=df['Speed Limit (km/h)'].apply(speed_category)
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df,x='Speed Risk Category',palette=["yellow","red","green"])
plt.title("Accidents by Speed Risk Category")
plt.xlabel("Speed Category")
plt.ylabel("Number of Accidents")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df,x='Speed Risk Category',hue='Accident Severity',palette=["red","green","orange"])
plt.title("Accident Severity based on Speed Risk Category")
plt.xlabel("Speed Risk Category")
plt.ylabel("Accident Count")
plt.show()
```

The chart clearly shows that high-speed environments contribute significantly to serious or fatal accidents. Medium-speed zones also show a notable number of accidents due to more active traffic, while low-speed areas have fewer severe incidents.

**Visualization: Accidents per Year (Trend Analysis)**

```python
plt.figure(figsize=(10,6))
year_counts=df['Year'].value_counts().sort_index()
sns.lineplot(x=year_counts.index,y=year_counts.values,marker="o",color="red")
plt.title("Yearly Trend of Road Accidents in India")
plt.xlabel("Year")
plt.ylabel("Number of Accidents")
plt.grid(True)
plt.show()
```

A clear trend can be observed in the yearly accident occurrences. The number of recorded accidents fluctuates due to factors like better road infrastructure, stricter traffic rules, pandemic lockdowns, and improvements in vehicle safety. An increasing trend symbolizes the need for further safety awareness and road monitoring.

**Visualization: Correlation Heatmap**

```python
#Select only numeric columns for corrleation
numeric_df=df.select_dtypes(include=[np.number])
plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(),annot=True,cmap="YlOrRd",linewidth=0.5) #YlOrRd --> Yellow,Orange,Red
plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```

Features with low correlation indicate that they may not significantly impact accident severity.
